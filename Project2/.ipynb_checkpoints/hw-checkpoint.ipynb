```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MinMaxScaler

from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score, KFold
from sklearn.metrics import confusion_matrix, make_scorer, accuracy_score, precision_score, recall_score, f1_score 
```

```python
data  = pd.read_csv("credit_cards.csv")
data.head()
```

```python
data = data.drop(["ID"], axis=1)
data.head()
```

```python
categorical_index = [1, 2, 3, 5, 6, 7, 8, 9, 10]
categorical_feature = []
for i in categorical_index:
    row = data.iloc[:, i].values.T
    if np.min(row) < 0: row += abs(np.min(row))
    print(np.unique(row))
    categorical_feature.append(data.iloc[:, i].values.T)
categorical_feature = np.vstack(categorical_feature)
categorical_feature = categorical_feature.T
```

```python
one_hot_encoder = OneHotEncoder()
categorical_feature = one_hot_encoder.fit_transform(categorical_feature)
categorical_feature = categorical_feature.todense()
```

```python
numerical_feature = []
for i in range(data.shape[1] - 1):
    if i not in categorical_index:
        numerical_feature.append(data.iloc[:, i])
numerical_feature = np.vstack(numerical_feature)
numerical_feature = numerical_feature.T
numerical_feature.shape
```

```python
scaler = MinMaxScaler()
normalized_numerical = scaler.fit_transform(numerical_feature)
```

```python
data_index = [i for i in range(data.shape[0])]
train_index, test_index = train_test_split(data_index, test_size=0.2)
```

## Task 1

```python
from sklearn.linear_model import LogisticRegression
```

```python
scoring = {'accuracy' : make_scorer(accuracy_score), 
           'precision' : make_scorer(precision_score),
           'recall' : make_scorer(recall_score), 
           'f1_score' : make_scorer(f1_score)}
kfold = KFold(n_splits=5, random_state=0)
```

```python
#normalized numerical
feature = np.concatenate((categorical_feature, normalized_numerical), axis=1)
label = data.iloc[:, -1]
####
y_test = label[test_index]
X_test = feature[test_index]
clf = LogisticRegression().fit(X_test, y_test)
pred = clf.predict(X_test)

#clf = LogisticRegression() 

accuracy = cross_val_score(clf, feature, label, cv=5, scoring="accuracy")
f1 = cross_val_score(clf, feature, label, cv=5, scoring="f1")
recall = cross_val_score(clf, feature, label, cv=5, scoring="recall")
precision = cross_val_score(clf, feature, label, cv=5, scoring="precision")
```

```python
confusion_matrix(y_test, pred)
```

```python
accuracy, f1, recall, precision
```

$$
\text { Confusion Matrix }={\text {A Matrix including True Positive, False Positive, False Positive, True Negative}}
$$

$$
\text { Accuracy }=\frac{\text {True Positive+False Negative}}{\text {True Positive+False Positive+False Positive+True Negative}}
$$

$$
\text { Precision }=\frac{\text {True Positive}}{\text {True Positive+False Positive}}
$$

$$
\text { Recall }=\frac{\text {True Positive}}{\text {True Positive}+\text {False Negative}}
$$

$$
\mathrm{F} 1=2 \times \frac{\text { Precision } * \text {Recall}}{\text {Precision}+\text {Recall}}
$$

```python
# - Oversampling
# - Data synthesis
from imblearn.over_sampling import RandomOverSampler
```

```python
#normalized numerical
feature = np.concatenate((categorical_feature, normalized_numerical), axis=1)
label = data.iloc[:, -1]

ros = RandomOverSampler(random_state=0)
X_train = feature[train_index]
y_train = label[train_index]
X_test = feature[test_index]
y_test = label[test_index]

X_train_resample, y_train_resample = ros.fit_resample(X_train, y_train)
clf = LogisticRegression() 
clf.fit(X_train_resample, y_train_resample)
pred = clf.predict(X_test)
```

```python
confusion_matrix(y_test, pred)
```

```python
accuracy_score(y_test, pred), f1_score(y_test, pred), recall_score(y_test, pred), precision_score(y_test, pred)
```
